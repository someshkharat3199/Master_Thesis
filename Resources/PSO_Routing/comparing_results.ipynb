```python
def read_data(file_name):
  with open("/content/gdrive/MyDrive/Networks/"+file_name,'rb') as file:
    data = pickle.load(file)
    file.close()
    return data
```

```python
PSO_ad2_route_optimality_ratio = read_data("PSO_ad2_route_optimality_ratio.pkl")
PSO_ad2_route_failure_ratio= read_data("PSO_ad2_route_failure_ratio.pkl")
PSO_ad2_results = read_data("PSO_ad2_results.pkl")
PSO_ad2_percent_gap = read_data("PSO_ad2_percent_gap.pkl")
PSO_ad2_iterations_to_converge = read_data("PSO_ad2_iterations_to_converge.pkl")
PSO_ad2_time_to_converge = read_data("PSO_ad2_time_to_converge.pkl")
```

```python
PSO_ad2_near_route_optimality_ratio = {}
PSO_ad2_near_route_failure_ratio = {}
for n_device in PSO_ad2_percent_gap.keys():
  near = 0
  if len(PSO_ad2_percent_gap[n_device]) == 0:
    print(n_device)
    continue
  for percent in PSO_ad2_percent_gap[n_device]:
    if percent <= 4:
      near = near + 1
  PSO_ad2_near_route_optimality_ratio[n_device] = near/len(PSO_ad2_percent_gap[n_device])
  PSO_ad2_near_route_failure_ratio[n_device] = (len(PSO_ad2_percent_gap[n_device])-near)/len(PSO_ad2_percent_gap[n_device])
```
